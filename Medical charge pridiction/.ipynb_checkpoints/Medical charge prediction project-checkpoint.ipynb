## spliting features and target

```python
x = df.drop(columns='charges')
y = df['charges'].copy()
```

## creating a full pipeline

```python
# sk.pipeline.pipeline([("scaling")])
```

## spliting the dataset

```python
xtrain,xtest, ytrain, ytest = sk.model_selection.train_test_split(x,y,random_state=42,shuffle=True)
```

```python
l = [xtrain,xtest, ytrain, ytest]
for i in l:
    print(i.head(2),"\n")  
```

```python
scaler = sk.preprocessing.MinMaxScaler()
```

```python
cols_to_scale = ['age','bmi','children']
cols_to_label = ['sex','smoker','region']
```

```python
scaler.fit(xtrain[cols_to_scale])
xtrain[cols_to_scale] = scaler.transform(xtrain[cols_to_scale])
xtest[cols_to_scale] = scaler.transform(xtest[cols_to_scale])
```

```python
xtrain = pd.get_dummies(xtrain,columns=cols_to_label,drop_first=True)
xtest = pd.get_dummies(xtest,columns=cols_to_label,drop_first=True)
```

```python
print(xtrain.head().T)
print(xtest.head().T)
```

## prediction and model fitting

```python
lr = sk.linear_model.LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
pred = lr.predict(xtest)
```

```python
sk.metrics.root_mean_squared_error(ytest,pred)
```

```python
sns.histplot(x=df['charges'],bins=30)
```

```python

sk.metrics.r2_score(ytest,pred)
```

## spliting the data for train and test

```python
xtrain,xtest = x.iloc[:60000], x.iloc[60000:]
ytrain,ytest = y.iloc[:60000], y.iloc[60000:]
```

```python
only_y_train = only_y.iloc[:60000]
only_y_test = only_y.iloc[60000:]
```

```python
only_y
```

## scaling the data

```python
scaler = sk.preprocessing.MinMaxScaler()
scaler.fit(xtrain)
xtrain = scaler.transform(xtrain)
```

## fitting model

```python
model = sk.linear_model.SGDClassifier()
```

```python
model.fit(xtrain,only_y_train)
```

```python
pred = model.predict(scaler.transform(xtest))
```

```python
sk.metrics.accuracy_score(only_y_test,pred)*100
```

```python
cm = sk.metrics.confusion_matrix(only_y_test,pred)
```
